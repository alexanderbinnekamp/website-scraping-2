```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from datetime import date
```

```python
dictionary = {'Title_edition' : [], 'Date_of_publication' : [], 'Link' : []}


current_year = date.today().year

x = range(int(current_year) - 1997)
for i in x:
    i = i + 1997
    URL = 'https://www.economist.com/weeklyedition/archive?year=' + str(i)

    html_text = requests.get(URL).text
    soup = BeautifulSoup(html_text, 'lxml')

    editions = soup.find_all('div', class_ = 'edition-teaser__text')

    for edition in editions:

        dictionary['Title_edition'].append(edition.span.text)
        dictionary['Date_of_publication'].append(edition.time.text)

        link = edition.h2.a['href']
        link = "https://www.economist.com/" + link
        dictionary['Link'].append(link)

dataframe = pd.DataFrame(dictionary)
dataframe.to_csv('C:/Users/alexa/OneDrive/Documenten/Collecting Data & Tools and Methods/economist.csv')

print(dictionary)
```
